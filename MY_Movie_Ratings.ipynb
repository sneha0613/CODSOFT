```python
df_movie['Rating'] = pd.to_numeric(df_movie['Rating'], errors='coerce')
def categorized_rating(rating):
  if(rating>8.0):
    return 'Excellent'
  elif(6.0<rating<8.0):
    return 'Good'
  elif(5.0<rating<6.0):
    return 'Average'
  else:
    return 'bad'
```

```python
df_movie['categorized_rating']=df_movie['Rating'].apply(categorized_rating)
```
